```python
import matplotlib.pyplot as plt
import numpy as np

four2016 = df[(df.mag >= 4) & (df.mag < 4.5) & (df.time >= '2016-01-01T00:00:00') & (df.time <= '2016-12-31T23:59:59')]
four2017 = df[(df.mag >= 4) & (df.mag < 4.5) & (df.time >= '2017-01-01T00:00:00') & (df.time <= '2017-12-31T23:59:59')]
four2018 = df[(df.mag >= 4) & (df.mag < 4.5) & (df.time >= '2018-01-01T00:00:00') & (df.time <= '2018-12-31T23:59:59')]
four2019 = df[(df.mag >= 4) & (df.mag < 4.5) & (df.time >= '2019-01-01T00:00:00') & (df.time <= '2019-12-31T23:59:59')]

five2016 = df[(df.mag >= 4.5) & (df.mag < 5) & (df.time >= '2016-01-01T00:00:00') & (df.time <= '2016-12-31T23:59:59')]
five2017 = df[(df.mag >= 4.5) & (df.mag < 5) & (df.time >= '2017-01-01T00:00:00') & (df.time <= '2017-12-31T23:59:59')]
five2018 = df[(df.mag >= 4.5) & (df.mag < 5) & (df.time >= '2018-01-01T00:00:00') & (df.time <= '2018-12-31T23:59:59')]
five2019 = df[(df.mag >= 4.5) & (df.mag < 5) & (df.time >= '2019-01-01T00:00:00') & (df.time <= '2019-12-31T23:59:59')]

six2016 = df[(df.mag >= 5) & (df.mag < 6) & (df.time >= '2016-01-01T00:00:00') & (df.time <= '2016-12-31T23:59:59')]
six2017 = df[(df.mag >= 5) & (df.mag < 6) & (df.time >= '2017-01-01T00:00:00') & (df.time <= '2017-12-31T23:59:59')]
six2018 = df[(df.mag >= 5) & (df.mag < 6) & (df.time >= '2018-01-01T00:00:00') & (df.time <= '2018-12-31T23:59:59')]
six2019 = df[(df.mag >= 5) & (df.mag < 6) & (df.time >= '2019-01-01T00:00:00') & (df.time <= '2019-12-31T23:59:59')]


seven2016 = df[(df.mag >= 6) & (df.mag < 7) & (df.time >= '2016-01-01T00:00:00') & (df.time <= '2016-12-31T23:59:59')]
seven2017 = df[(df.mag >= 6) & (df.mag < 7) & (df.time >= '2017-01-01T00:00:00') & (df.time <= '2017-12-31T23:59:59')]
seven2018 = df[(df.mag >= 6) & (df.mag < 7) & (df.time >= '2018-01-01T00:00:00') & (df.time <= '2018-12-31T23:59:59')]
seven2019 = df[(df.mag >= 6) & (df.mag < 7) & (df.time >= '2019-01-01T00:00:00') & (df.time <= '2019-12-31T23:59:59')]

max2016 = df[(df.mag >= 7) & (df.mag <= 10) & (df.time >= '2016-01-01T00:00:00') & (df.time <= '2016-12-31T23:59:59')]
max2017 = df[(df.mag >= 7) & (df.mag <= 10) & (df.time >= '2017-01-01T00:00:00') & (df.time <= '2017-12-31T23:59:59')]
max2018 = df[(df.mag >= 7) & (df.mag <= 10) & (df.time >= '2018-01-01T00:00:00') & (df.time <= '2018-12-31T23:59:59')]
max2019 = df[(df.mag >= 7) & (df.mag <= 10) & (df.time >= '2019-01-01T00:00:00') & (df.time <= '2019-12-31T23:59:59')]
```

```python
barWidth = 0.25

bars1 = [four2016.shape[0], four2017.shape[0], four2018.shape[0], four2019.shape[0]]
bars2 = [five2016.shape[0], five2017.shape[0], five2018.shape[0], five2019.shape[0]]
bars3 = [six2016.shape[0], six2017.shape[0], six2018.shape[0], six2019.shape[0]]
bars4 = [seven2016.shape[0], seven2017.shape[0], seven2018.shape[0], seven2019.shape[0]]
bars5 = [max2016.shape[0], max2017.shape[0], max2018.shape[0], max2019.shape[0]]


r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

# Make the plot
plt.bar(r1, bars1, color='black', width=barWidth, edgecolor='white', label='Group 1')
plt.bar(r2, bars2, color='purple', width=barWidth, edgecolor='white', label='Group 2')
plt.bar(r3, bars3, color='orange', width=barWidth, edgecolor='white', label='Group 3')
plt.bar(r4, bars4, color='blue', width=barWidth, edgecolor='white', label='Group 4')

# Add xticks on the middle of the group bars
plt.xlabel('Year', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```
